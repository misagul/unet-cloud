```python
import cv2
import numpy as np
from IPython.display import Image
from PIL import Image as PilImage
```

```python
g = cv2.imread("g-0.png")
p = cv2.imread("p-0.png")
f = np.concatenate((g,p), axis=1)
cv2.imwrite("f.png", f)
Image("f.png")
```

```python
l_red = np.array([217, 136, 128])
l_green = np.array([130, 224, 170])

d_red = np.array([100, 30, 22])
d_green = np.array([20, 90, 50])

red = np.array([255, 0, 0])
green = np.array([0, 255, 0])

w = np.array([255, 255, 255])
b = np.array([0, 0, 0])
```

```python
new = np.zeros([384,384,3], dtype=int)
total = 384*384
cloud_correct = 0
cloud_mistake = 0
not_cloud_correct = 0
not_cloud_mistake = 0

for i in range(0,384):
    for j in range(0,384):
        if np.array_equal(g[i][j], w):
            if np.array_equal(p[i][j], w):
                new[i][j] = l_green
                cloud_correct += 1
            else:
                new[i][j] = d_green
                cloud_mistake += 1
        else:
            if np.array_equal(p[i][j], b):
                new[i][j] = l_red
                not_cloud_correct += 1
            else:
                new[i][j] = d_red
                not_cloud_mistake += 1
im = PilImage.fromarray((new * 1).astype(np.uint8)).convert('RGB')
im.save("res.png")

print(f"Total:\t\t\t{total}\nCloud Correct:\t\t{cloud_correct}\nCloud Mistake:\t\t{cloud_mistake}\nNot Cloud Correct:\t{not_cloud_correct}\nNot Cloud Mistake:\t{not_cloud_mistake}\nRate:\t\t\t{(cloud_correct+not_cloud_correct)/total:.2f}")
```
